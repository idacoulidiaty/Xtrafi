```python
def load_excel_data_dynamic_start(file_path):
    # Lire le fichier sans parser les colonnes pour chercher la ligne clé
    preview_df = pd.read_excel(file_path,header=None, engine= "calamine")

    # Chercher la ligne contenant le texte cible
    start_row_idx = preview_df[0].astype(str).str.contains("Liste des données qui seront utilisées pour", case=False, na=False)

    if start_row_idx.any():
        # Trouver l'index de la ligne et +1 pour commencer après
        start_index = start_row_idx[start_row_idx].index[0] + 1
        df = pd.read_excel(file_path, skiprows=start_index, engine="calamine")
        return df
    else:
        raise ValueError("❌ Ligne de début de données introuvable dans le fichier.")
```

```python
file_path = "./data/restitution-30-20250312161809.xls"
df_onglet_1 = load_excel_data_dynamic_start(file_path)
df_onglet_1
```
